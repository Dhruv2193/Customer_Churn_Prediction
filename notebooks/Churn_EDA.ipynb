```python
# CUSTOMER CHURN PREDICTION WITH EDA

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# -----------------------------
# 1. LOAD DATA
# -----------------------------
df = pd.read_csv("Churn.csv")

# -----------------------------
# 2. BASIC INFO
# -----------------------------
print("\nData shape:", df.shape)
print("\nMissing values:\n", df.isnull().sum())
print("\nChurn distribution:\n", df['Closed'].value_counts(normalize=True))

# -----------------------------
# 3. EDA
# -----------------------------

# Churn count plot
plt.figure(figsize=(5,4))
sns.countplot(x='Closed', data=df, palette='coolwarm')
plt.title("Churn Distribution (0 = Retained, 1 = Churned)")
plt.show()

# Churn by Gender
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', hue='Closed', data=df, palette='Set2')
plt.title("Churn by Gender")
plt.show()

# Churn by Geography
plt.figure(figsize=(6,4))
sns.countplot(x='Geography', hue='Closed', data=df, palette='Set1')
plt.title("Churn by Geography")
plt.show()

# Distribution of Credit Score by Churn
plt.figure(figsize=(6,4))
sns.histplot(df, x='Credit Score', hue='Closed', kde=True, palette='husl')
plt.title("Credit Score Distribution by Churn")
plt.show()

# Correlation heatmap (numerical features)
# Correlation heatmap (numeric features only)
plt.figure(figsize=(8,6))
numeric_df = df.select_dtypes(include=[np.number])  # select only numeric columns
sns.heatmap(numeric_df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap (Numeric Features)")
plt.show()



```
